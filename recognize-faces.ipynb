```python
front_cascade_fn='opencv-2.4.9/data/haarcascades/haarcascade_frontalface_alt2.xml'
side_cascade_fn ='opencv-2.4.9/data/haarcascades/haarcascade_profileface.xml'

scaleFactor=1.3
minNeighbors=2
minSize=(20, 20)
flags=cv2.cv.CV_HAAR_SCALE_IMAGE 
```

```python
front_cascade = cv2.CascadeClassifier(front_cascade_fn)
side_cascade  = cv2.CascadeClassifier(side_cascade_fn)
```

```python
def detect_faces( filename, cascade ):

    try:
    
        img = cv2.imread( filename, cv2.IMREAD_GRAYSCALE )
        
        mindim  = min( img.shape )
        minSize = ( int( .025 * mindim ), int( .025 * mindim ) )
    
        rects = cascade.detectMultiScale(img, scaleFactor=scaleFactor,
                                         minNeighbors=minNeighbors,
                                         minSize=minSize, flags=flags)
        
        return rects
    
    except:
        pass
        
    return np.nan
```

```python
detect_faces( user_df.iloc[1]['filename'], cascade=side_cascade )
```

```python
user_df['front_faces'] = user_df['filename'].apply( detect_faces, cascade=front_cascade )
```

```python
user_df['side_faces'] = user_df['filename'].apply( detect_faces, cascade=side_cascade )
```

```python
def count_faces( rects ):
    
    if type( rects ) != np.ndarray:
        return 0
    
    return np.atleast_2d( rects ).shape[0]
```

```python
user_df['num_front_faces'] = user_df['front_faces'].apply( count_faces )
user_df['num_side_faces'] = user_df['side_faces'].apply( count_faces )
```

```python
user_df['num_faces'] = user_df['num_front_faces'] + user_df['num_side_faces']
```

```python
def display_faces( row ):
    
    img = cv2.imread( row['filename'], cv2.IMREAD_GRAYSCALE )
    
    plt.imshow( img )
    currentAxis = plt.gca()
    
    rects = row['front_faces']
    if type( rects ) == np.ndarray:
        for i in range( rects.shape[0] ):
            currentAxis.add_patch( Rectangle( (rects[i,0] - .5, 
                                               rects[i,1] - .5), 
                                               rects[i,2], 
                                               rects[i,3], edgecolor="red", facecolor="none"))
        
    rects = row['side_faces']
    if type( rects ) == np.ndarray:
        for i in range( rects.shape[0] ):
            currentAxis.add_patch( Rectangle( (rects[i,0] - .5, 
                                               rects[i,1] - .5), 
                                               rects[i,2], 
                                               rects[i,3], edgecolor="lightblue", facecolor="none"))
```
