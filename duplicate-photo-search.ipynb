# Train vocabulary

```python
class BOWKMeansTrainer( object ):
    
    def __init__( self, clusterCount=1000 ):
        
        self.clusterCount = clusterCount
        self.descriptors  = []
        
    def add( self, desc ):
        
        self.descriptors += [ desc ]
        
    def descriptorCount( self ):
        
        return len( self.descriptors )
    
    def cluster( self ):
        
        from sklearn import cluster
    
        # create descriptor array
        desc_array = sp.vstack( tuple(self.descriptors) ).astype( 'double' )
    
        # create a k-means of vsize clusters
#        self.centroids, labels, inertia = cluster.k_means( flist, self.vsize )
        mbkm = cluster.MiniBatchKMeans( self.clusterCount )

        # fit to the set of provided feature descriptors
        mbkm.fit( desc_array )

        # retain the centroids the centroids
        return mbkm.cluster_centers_
```

# Quantize descriptors to vocabulary

```python
class BOWImgDescriptorExtractor(object):
    
    def __init__( self, dextractor ):
        
        self.dextractor = dextractor
        
    def setVocabulary( self, vocabulary ):
        
        self.vocabulary = vocabulary

    def compute_from_image( self, image, keypoints ):
            
        # get the feature descriptions
        desc = dextractor.compute( image, keypoints )

        return self.compute( desc )
        
    def compute( self, desc ):
        
        from sklearn import metrics

        vsize = self.vocabulary.shape[0]

        # compute the distance between the vocabulary centers and the desciptors
        dist = metrics.pairwise_distances( self.vocabulary, desc )

        # quantize to the closest cluster
        nearest_centroids = sp.argmin( dist, axis=0 )
        
        # generate histogram given that
        hist, dummy = sp.histogram( nearest_centroids, vsize, [0, vsize] )
        return hist
    
    
```

```python
from sklearn import metrics

# compute the distance between the vocabulary centers and the desciptors
dist = metrics.pairwise_distances( desc_array, search )
```

```python
idx = sp.argmin( dist, axis=0 )[0]
idx
```

```python
s_res_kp = room_df['keypoints'].iloc[idx]
s_res_de = room_df['descriptors'].iloc[idx]
```

```python
raw_matches = matcher.knnMatch(desc, trainDescriptors = s_res_de, k = 2)
```

```python
def filter_matches(kp1, kp2, matches, ratio = 0.75):
    mkp1, mkp2 = [], []
    for m in matches:
        if len(m) == 2 and m[0].distance < m[1].distance * ratio:
            m = m[0]
            mkp1.append( kp1[m.queryIdx] )
            mkp2.append( kp2[m.trainIdx] )
    p1 = np.float32([kp.pt for kp in mkp1])
    p2 = np.float32([kp.pt for kp in mkp2])
    kp_pairs = zip(mkp1, mkp2)
    return p1, p2, kp_pairs
```

```python
p1, p2, pairs = filter_matches( kp, s_res_kp, raw_matches )
```
